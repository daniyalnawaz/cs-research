```python
import http.client, urllib.parse, csv
from bs4 import BeautifulSoup
import pandas as pd
```

Create a date range:

```python
datelist = pd.date_range(pd.datetime(2015, 1, 1), pd.datetime(2015,8,25)).tolist()
```

Create a function that returns the HTML:

```python
def getHTML(m, d, y):
    print('Date: ', y, m, d)
    params = urllib.parse.urlencode({'sday': d, 'mth': m, 'yr': y})
    headers = {"Content-type": "application/x-www-form-urlencoded", "Accept": "text/plain"}

    conn = http.client.HTTPConnection("dps.kse.com.pk:80")

    conn.request("POST", "/webpages/history_byDate1.php", params, headers)
    response = conn.getresponse()
    print('---', response.status, response.reason)
    
    return response.read()
```

Function to get content from a  object safely:

```python
def getVal(dom):
    if (len(dom.contents) == 0):
        return ""
    
    val = dom.contents[0]
    
    val = val.replace(",", "")
    
    if val == None:
        return ""
    return val
```

Create an empty document:

```python
f = open("data/2015.csv", 'a')
wr = csv.writer(f)

wr.writerow([
    'date',
    'symbol',
    'ldcp',
    'open',
    'high',
    'low',
    'closing',
    'turnover'
])
```

```python
for date in datelist:
    html = getHTML("{0:0=2d}".format(date.month), "{0:0=2d}".format(date.day), date.year)
    soup = BeautifulSoup(html, "html.parser")
    rows = soup.find_all('tr')
    count = 0
    
    for row in rows:
        tds = row.find_all('td')
        if len(tds) != 7:
            continue
    
        link = tds[0].find('a')
        if link == None:
            continue
        
        wr.writerow([
            date.strftime("%Y-%m-%d"),
            getVal(link),
            getVal(tds[1]),
            getVal(tds[2]),
            getVal(tds[3]),
            getVal(tds[4]),
            getVal(tds[5]),
            getVal(tds[6])
        ])
        
        count = count + 1
    
    print('---', 'Total Values:', count)

f.close()
```
